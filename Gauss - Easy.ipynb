#### Hierbei muss nur die Anzahl zu berechnenden Unbekannten und die jeweiligen Koeffizienten zeilenweise von Links nach Rechts eingegeben wurde. 

#### Daraufhin wird schrittweise die obere Dreiecksmatrix berechnet und das Ergebnis ausgegeben

```python
#Import benötigter Bibliotheken
import numpy as np
import sys

# Anzahl Unbekannten
n = int(input('Anzahl Unbekannte: '))

# Erstelle Numpy-Arrays
a = np.zeros((n,n+1))
x = np.zeros(n)

# Eingabe Koeffizienten
print('Koeffizienten:')
for i in range(n):
    for j in range(n+1):
        a[i][j] = float(input( 'a['+str(i)+']['+ str(j)+']='))

# Gauß‘sches Eliminationsverfahren
for i in range(n):
    if a[i][i] == 0.0:
        sys.exit('Sonderfall - Keine oder unendlich viele Lösungen')
        
    for j in range(i+1, n):
        ratio = a[j][i]/a[i][i]
        
        for k in range(n+1):
            a[j][k] = a[j][k] - ratio * a[i][k]
    print(a)
        
# Rücksubstitution
x[n-1] = a[n-1][n]/a[n-1][n-1]

for i in range(n-2,-1,-1):
    x[i] = a[i][n]
    
    for j in range(i+1,n):
        x[i] = x[i] - a[i][j]*x[j]
    
    x[i] = x[i]/a[i][i]


# Lösung
print('\n Lösung: ')
for i in range(n):
    print('X%d = %0.2f' %(i+1,x[i]), end = '\t')
```
